```python
def remove_outliers(col):
    sorted(col)
    q1,q3=np.nanpercentile(col,[25,75])
    IQR=q3-q1
    ul=q3+(1.5*IQR)
    ll=q1-(1.5*IQR)
    return(ul,ll)
```

```python
for i in df_x.columns:
    ul,ll=remove_outliers(df_x[i])
    df_x[i]=np.where(df_x[i]>ul, ul, df_x[i])
    df_x[i]=np.where(df_x[i]<ll, ll, df_x[i])
```

```python
# as Logistic regression is sensistive to multicolinearity hence we need to drop varibales with high correlations
# here we have used Vif(Variation INflation Factor) in order to treat multicolinearity in our dataset
```

```python
# lets check for multicolinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(X):
    vif=pd.DataFrame() # creating dataframe for storing variables and its values
    vif['Variables']=X.columns
    vif['Values']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    return(vif)
```
