# Analysing Return 

```python
df_returns = np.log(df.drop(['Date','dates'],axis=1)).diff(axis = 0, periods = 1) 
```

```python
df_returns
```

# Checking the rows & columns of datases

```python
df_returns.shape
```

*checking top 5 rows*

```python
df_returns.head()
```

# Calculating stock means

```python
df_means = df_returns.mean(axis = 0)
df_means
```

# Calculating stock Standard Deviation

```python
df_std = df_returns.std(axis = 0)
df_std
```

```python
df_summary = pd.DataFrame({'Average':df_means, 'Volatility': df_std})
df_summary
```

# Let us plot & see what they say about stock prices

```python
plt.scatter(df_std, df_means, edgecolors='r')
plt.axhline(y=0,linestyle='--', color = "red")
plt.xlabel('Volatility')
plt.ylabel('Average')
plt.show()
```

```python
df_summary[df_summary['Average'] > 0].sort_values(by = 'Volatility')
```
